## <font color=green>Key Concepts in Classification :<font>

### 1.Classes : 
        The predifined categories or labels that data points can belong to.For example ,in a spam email classifier,the classes would be "spam" and "not spam"
### 2.Features :
        The input variables (or predictors) that help determine the class of each data point .For instance ,words in an email might be features in a spam classifier.
### 3.Labels :
        The actual categories assigned to each data point in the training data,which help the model learn the correct class during training.

## <font color=green>Types of Classification :<font>
#### 1.Binary Classification: 
        Classifies data into one of two classes. Examples include spam detection ("spam" or "not spam") and medical diagnosis (positive or negative for a disease).

#### 2.Multiclass Classification: 
        Classifies data into one of three or more classes. Examples include categorizing images of animals into "cat," "dog," and "bird" classes.

#### 3.Multilabel Classification: 
        Allows multiple labels for each instance. For example, an image could be labeled as "cat" and "outdoor" simultaneously.

## <font color='green'>Common Classification Algorithms :<font>
#### 1.Logistic Regression: 
        A statistical model that predicts the probability of a binary class. It uses a sigmoid function to produce probabilities that can be mapped to classes.

#### 2.K-Nearest Neighbors (KNN): 
        A non-parametric algorithm that assigns a class to a data point based on the majority class of its nearest neighbors.

#### 3.Support Vector Machine (SVM): 
        Finds the optimal hyperplane that maximizes the margin between classes, used mainly in binary classification.

#### 4.Decision Trees: 
        Use a tree-like structure where each node represents a feature, and each branch represents a decision rule, helping to classify the data based on feature splits.

#### 5.Random Forest: 
        An ensemble method that combines multiple decision trees to improve accuracy and reduce overfitting.

#### 6.Naive Bayes: 
        A probabilistic classifier based on Bayes' Theorem, assuming independence among predictors. It works well in text classification tasks.

#### 7.Neural Networks: 
        Used for complex tasks like image or speech recognition, where deep learning models (with multiple layers) learn intricate patterns in the data.

## <font color='green'>Classification Performance Metrics:<font>
#### 1.Accuracy: 
        The percentage of correctly classified instances out of the total instances. It is effective when classes are balanced.

#### 2.Precision: 
        The proportion of true positive predictions among all positive predictions. Useful for evaluating how many selected instances are relevant, especially in imbalanced classes.

#### 3.Recall (Sensitivity): 
        The proportion of true positive predictions among all actual positives, measuring the model's ability to capture all relevant instances.

#### 4.F1 Score: 
        The harmonic mean of precision and recall, balancing the two. It’s useful when classes are imbalanced.

#### 5.Confusion Matrix: 
        A table that summarizes the true positive, false positive, true negative, and false negative values, providing detailed insights into the performance for each class.

## <font color='blue'>Implementation of Confusion Matrix for Binary classification using Python<font>

#### Step 1: Import the necessary libraries

```python
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
import seaborn as sns
import matplotlib.pyplot as plt 
```

#### Step 2: Create the NumPy array for actual and predicted labels

```python
actual    = np.array(
  ['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog'])
predicted = np.array(
  ['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])
```

#### Step 3: Compute the confusion matrix

```python
cm = confusion_matrix(actual,predicted)
```

#### Step 4: Plot the confusion matrix with the help of the seaborn heatmap

```python
sns.heatmap(cm, 
            annot=True,
            fmt='g', 
            xticklabels=['Dog','Not Dog'],
            yticklabels=['Dog','Not Dog'])
plt.ylabel('Actual', fontsize=13)
plt.title('Confusion Matrix', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.xlabel('Prediction', fontsize=13)
plt.gca().xaxis.tick_top()

plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Prediction', ha='center', fontsize=13)
plt.show()
```

#### Step 5: Classifications Report based on Confusion Metrics

```python
print(classification_report(actual, predicted))
```

## <font color='green'>Implementation of Confusion Matrix for Multi-Class classification using Python<font>

#### Step 1: Import the necessary libraries

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
```

#### Step 2: Create the NumPy array for actual and predicted labels

```python
# Actual labels
y_true = ['Cat'] * 10 + ['Dog'] * 12 + ['Horse'] * 10

# Predicted labels 
y_pred = ['Cat'] * 8 + ['Dog'] + ['Horse'] + ['Cat'] * 2 + ['Dog'] * 10 + ['Horse'] * 8 + ['Dog'] * 2

# Classes
classes = ['Cat', 'Dog', 'Horse']
```

#### Step 3: Generate and Visualize the Confusion Matrix

```python
# Generate the confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix', fontsize=15, pad=20)
plt.xlabel('Prediction', fontsize=11)
plt.ylabel('Actual', fontsize=11)
#Customizations
plt.gca().xaxis.set_label_position('top')
plt.gca().xaxis.tick_top()
plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Prediction', ha='center', fontsize=13)

plt.show()
```

#### Step 4: Print the Classification Report

```python
# Print classification report
print(classification_report(y_true, y_pred, target_names=classes))
```


<h1><font color=darkblue> Correlation<font></h1>


        Correlation is a statistical measure that describes the degree to which two variables move in relation to ach other.It indicates both the directiin and strength of a relationship between variables,helping to identify patterns in data.

## <font color='green'>Types of Correlation<font>

#### 1.Positive Correlation :
        When one variable increases,the other also increases.For example,height and weight often have a positive correlation.
#### 2.Negative Correlation :
        When one variable increases ,the other decreses .For example as temparature decreases,heating usage may increase.
#### 3.No Correlation :
        No apparent relationship between the variables,meaning they do not move together in any predictable wasy.

## <font color='green'>Correlation Coefficient<font>

        The correlation coefficient(often denoted as r) is a numerical value between -1 and 1 that quantifies correlation:

        * r=1: Perfect positive correlation
        * r=-1 : Perfect negative correlation
        * r=0 : No correlation

        The most commonly used measure is 'Pearson's correlation coeffiient ,which claculates the linear relationship between two continuous variables.'

## <font color='green'>Interpreting the Value of r<font>


        * Close to 1: Strong positive relationship.
        * Close to -1: Strong nagative relationship.
        * Around 0 : Weak or no linear relationship.

## <font color='green'> How to check the correlation values:<font>

#### 1.	Understand the Range of Correlation Values:
        o	Correlation coefficients range from -1 to 1:
        	+1: Perfect positive correlation (as one variable increases, the other also increases).
        	-1: Perfect negative correlation (as one variable increases, the other decreases).
        	0: No correlation (no linear relationship).
#### 2.	Identify Strong Correlations:
        o	Look for correlation values close to +1 or -1 for strong relationships.
        o	Values between 0.5 and 1 (or -0.5 and -1) indicate a strong positive or negative correlation, respectively.
        o	Values between 0.3 and 0.5 (or -0.3 and -0.5) indicate a moderate correlation.
        o	Values between 0 and 0.3 indicate a weak correlation.
#### 3.	Diagonal Elements:
        o	The diagonal of the correlation matrix will always be 1 because it represents the correlation of each variable with itself.
#### 4.	Focus on Target Correlations:
        o	Check the correlation values between the target variable (e.g., TenYearCHD) and other features to identify relevant predictors


## <font color='green'> How to select Features Based on Correlation:<font>



#### 1.	Choose Features with Strong Correlation to the Target:
        o	Select features with a high absolute correlation with the target variable (e.g., |correlation| > 0.5).
        o	This indicates that these features have a strong linear relationship with the target.
#### 2.	Avoid Multicollinearity:
        o	Check for high correlations between independent features (e.g., |correlation| > 0.8).
        o	If two features are highly correlated with each other, keep one and drop the other to avoid multicollinearity, which can cause problems in some models (e.g., linear regression).
#### 3.	Domain Knowledge:
        o	Use domain knowledge to verify the significance of features. Some features may have a lower correlation but are still important based on the context of the data (e.g., medically significant indicators).
#### 4.	Use a Heatmap for Visualization:
        o	Use a heatmap to visually identify clusters of high correlation.
        o	Strongly correlated features can be grouped and analyzed for redundancy.


## <font color='green'> Applications<font>
