## What is Multiple Linear Regression?

        Multiple Linear Regression is one of the Important regression algorithm which models the linear regression between a single dependent continuous variable and more than one independent variable.
        The Multiple Linear Regression,that targets variable(Y) is a linear combination of multiple predictor variable X1,X2,X3,X4....Xn.Since it is an enhancements of simple linear regression .So the same is applied for the Multiple Linear Regression Equation.

        Multiple Linear Regression is closly related to a simple linear regression model with the difference in the number of independence variables.Whereas the simple linear regression model predicts the value of a dependent variable based on the value of single independent variable ,in multiple Linear Regression the value of a dependent variable is predicted based on more than one independent variables.

## Multiple Linear Regression Equation:

### Y= b0 + b1* X1 + b2* X2 + b3* X3 ……….... + bn* Xn 

        Y= Dependent Variable or predicted variable 
        b0 = It is the y-intercept
        b0,b1,b2,b3..= Coefficients of models  that represents the change in Y relative to a one-unit change in x1 and x2 respectively.
        x1,x2,x3,x4....= Various Independent /Feature variable.

## Assumption of Linear Regression:

### * Linearity :
        The relationship between dependent and independent variables should be linear.
### * Homoscedasticity
### * Multivariable Normality :
        Multiple regression assumes that the residuals are normally distributed. It should like bell like curve
### * Independence of errors
### * Lack of Multicollinearity : 
        It is assumed that there is little or no multicollinearity in the data.Multicollinearity occurs when the features (or independent variables) are not independent of each other. that means no two independent variables should not be same features.
        

## Dummy Variable

        Using categorical data in Multiple Regression Models is a Powerful method to include non-numeric data types into a regression model Categorical data refers to data values which represent categories data values with a fixed and unordered number of values,for instance,gender(male/female).In a regression model,these values can be represented by dummy varibles-variables containing values such as 1 or 0 representing the presence or absence of the categorical.

## Dummy Variable Trap

        The Dummy Variable trap is a scenario in which two or more variables are highly correlated; in a simple terms,one variable can be predicted from the others.Intutively,there is a duplicate category:If we dropped the male category it is inherently defined in the female category(zero female value indicates male,and vice verse.) The solution to the dummy variable trap is to drop one of the categorical variables-if there are m number of categories ,use m-1 in the model,the value left out can be thought of as the reference value.

## 2.Backward Elimination:

        In Backward Elimination is a stepwise regression technique used to select the most significant variables in a Multiple Linear Regression(MLR) model.This process helps in building a simplified ,more interpretable model by removing features that do not significantly contribute to the prediction.

### Step 1: 
        Select the significance level to say in the model (e.g., SL=0.05)
### Step 2:
        Fit the full model with all possible predictors
### Step 3.
        Considered the predictors with the highest P-values. If P>SL ,go to step 4 ,otherwise go to FIN
### Step 4: 
        Remove the predictor
### step 5:
        Fit the model without the predictor

        And so after STEP-5,we go back to STEP-3 once agiain,we look for the variable with the highest value in our new model.
        We take it out remove.
        Basically,the STEP-4 we remove it.We fit the model again with one less variable and so on and we keep doing that untill we come to a point where the even the variable of the highest P-value that P-value is still less than our significance level.
            So if the condition P is greater than SL is not correct,then we don't go to STEP-4 anymore.
            Then we go to FIN
            ###### FIN => OUR MODEL is READY
                                                           

## 3. Forward Selection

        Forward Selection is a stepwise regression technique used in Multiple Linear Regression(MLR) to select the most significant predictors incrementally.In this process,we start with no features in the model and add them one at a time based on their significance in predicting the dependent variable.This help builds a more interpretable model by only including important variables.

  ##  Steps in Forward Selection

### Step 1: Initialize with No Variables: 
        Start with an empty model(no features)
### Step 2: Choose a Significant Level:
        Set a significant leval(SL) ,such as 0.05 ,to decide if a variable should be added to the model.
### Step 3: Add the Most Significant Variable:
        * For each iteration,test each remaining variable to see which one would improve the model the most if added.
        * Use the p-value(or another evaluation metric) to decide which variables is the most significant predictor of the dependent variable when added to the current model.
### Step 4: Check the P-value:
        * If the p-value of the new variable is below the chosen significance level(SL),add it to the model.
        * If no variables have a p-value below SL,stop the process.
### Step 5: Repeat:
            * Continue adding variables one by one until no additional variables significantly improve the model.

        Label Encoding and One-Hot Encoding are two techniques for converting categorical data into a numerical format suitable for machine learning algorithms. While they both serve the purpose of encoding categorical variables, they differ in how they represent the categories, which can impact model performance and interpretability.

###  1.Label Encoding :
        Label encoding assigns each category in a feature an integer value.It is simple but can sometimes introduce problems in interpolation by machine learning algorithms ,as it implies an ordinal relationship (an order or ranking ) between categories.


#### Example:
        For a "Color" column:
        
       * Data: Red, Green, Blue
       * Encoded Values:
           * Red → 0
           * Green → 1
           * Blue → 2

### Pros and Cons of Label Encoding
#### Pros:

        Simple and memory-efficient since it replaces categories with single integer values.
        Works well with ordinal categorical variables (variables with a natural order like Low, Medium, High).
#### Cons:

        Introduces ordinal bias in non-ordinal data. For example, an algorithm may interpret Green (1) as being "between" Red (0) and Blue (2), which may not make sense for nominal categories like colors.

## 2. One-Hot Encoding
        One-hot encoding creates a new binary column for each unique category in a feature. Each row has a 1 in the column corresponding to its category and 0 elsewhere. This method avoids ordinal relationships by treating each category as independent.

### Example:
        For a "Color" column:

        Data: Red, Green, Blue
        Encoded Values:
        Red → [1, 0, 0]
        Green → [0, 1, 0]
        Blue → [0, 0, 1]
### Pros and Cons of One-Hot Encoding
#### Pros:

        No ordinal bias: Each category is treated as an independent binary feature.
        Better suited for nominal categorical variables (variables with no inherent order).
#### Cons:

        Increases dimensionality: For features with many unique categories, it can lead to high-dimensional data, increasing computational cost and memory usage.
        Sparsity: It can create a sparse matrix, with many zeros, especially when there are many categories.        

## Choosing Between Label Encoding and One-Hot Encoding
### Use Label Encoding for:

        Ordinal categorical features where the order is meaningful (e.g., "Low," "Medium," "High").
        Situations where memory and computational efficiency are crucial (e.g., limited resources).
### Use One-Hot Encoding for:

        Nominal categorical features where there is no natural order (e.g., "Red," "Green," "Blue").
        Algorithms that may interpret numeric values as ordered, like linear regression or distance-based algorithms.

```python
# importing libraries   
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd   
#importing datasets   
data_set= pd.read_csv('50_Startups.csv')   
X= data_set.iloc[:, 0].values   
#only getting R&D Column 
Y= data_set.iloc[:, 4].values  #only getting Profit Column 
# Splitting the dataset into training and test set.   
from sklearn.model_selection import train_test_split   
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= 0.2, 
random_state=0)   
X_train=np.array(X_train).reshape(-1, 1) 
X_test=np.array(X_test).reshape(-1, 1) 
#Fitting the MLR model to the training set:   
from sklearn.linear_model import LinearRegression   
regressor= LinearRegression()   
regressor.fit(X_train, Y_train)   
#Predicting the Test set result: 
Y_pred= regressor.predict(X_test)   
X_pred= regressor.predict(X_train) 
#Cheking the score : 
print('Train Score: ', regressor.score(X_train, Y_train))   
print('Test Score: ', regressor.score(X_test, Y_test))   
#Visualizing  the Training set results: 
plt.scatter(X_train,Y_train,color='red') 
plt.plot(X_train,X_pred,color='blue') 
plt.title('R&D Spend vs Profit (Training set)') 
plt.xlabel("R&D Spend") 
plt.ylabel('Profit') 
plt.show() 
#Visualizing  the Test set results 
plt.scatter(X_test,Y_test,color='red') 
plt.plot(X_train,X_pred,color='blue') 
plt.title('R&D Spend vs Profit (Test set)') 
plt.xlabel("R&D Spend") 
plt.ylabel('Profit') 
plt.show()
```
