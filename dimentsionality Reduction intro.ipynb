# <font color='#FF33A1'> What is Dimentionality Reduction?</font>

 #####      'Dimentionality Reduction' is technique in Machine Learning and Data Science to reduce the number of input variables or features in a dataset while preserving as much information as possible.It helps in simplifying model,reducing compuatation time and avoiding overfitting

#####  1.This is where dimentionality reduction techniques come into play.This is simply reducing the dimension of your feature set.
        For Example: If we have variables with multicollinearity ,there we have to remove the one variable which has less impact to model performance,there we can reduce the overfitting

        If we take Decision Trees and SVM ,these model can't perform well when we have to many features called high dimentionality like 1000D,2000D . from this we should reduce to 2d or 3d or upto 10d based on the dataset.
##### 2.This technique allows you to find a small set of most impactful features among a large number of features.

        But in some cases we have a chances to loose the important data, but maximum we can extract the possible features.
##### 3.With this small set of principle features,you can run your prediction algorithms easily with better accuaracy.

## Feature Selection Methods

            Feature selection methods use the statistical relationship of input variables to the output variable.The methods mainly find the correlations among the features and try to select the most independent features that have no colinearity.It selects the features with highest importance.

            It will checks the any of the variables  have forms multicollinearity, then we should remove one of them which has low impact or low variance.
            In this we find the variance for every independent variable and select the highest variance varibles among all.

<h2> <font color='#FF33A1'>1.Filter Methods</font></h2>

        In this method,each feature is ranked based on the some univariant metrics. Then it selects the highest -ranking features.

Some of the common filtering methods are--

 #####       1.Varience:
         Based on the highest variance of feature ,it will selects those features into new dataset. For example in 10 features ,in this we have 5 with highest variance like that.
#####        2.Chi-Square:
        It will used to find the correlation between categoical varibales ,Unlike numerical variables uses correlation like pearsoman correlation.
#####        3.Correlation coefficients :
        It is our normal correlation between variables ,select highest correlation with dependent variables.
#####        4.Information gain or mutual infomation :
        In Decision Trees,Based on the Information gain we select the features.

<h2><font color='#FF33A1'>2.Wrapper Methods</font>

                Wrapper methods selects the features based on the specific classifier performance.With a greedy appraoch,it evaluated all the possible combinations of features against the evaluation criterion

        For Example:  If we take any model either regression or classification,and we take features we simply evaluate the model,but we don't know which features are important in this case ,by using forward selection we select  feature sequentially and train model whichever the features gives highest accuracies those will keep and remaining will be ignore.

Some of the popular wrapper methods are--

        1.Genetic Algorithms
        2.Recursive Feature Elimination
        3.Sequential Feature Selection algorithms.
