# <font color='#FF33A1'>‚úÖ What is PCA?</font>

##### Principle Component Analysis ,or PCA is a dimensionality-reduction method that is often used to reduce the dimentionality of large data sets,by transforming a large set of varibles into a smaller one that still contains most of the information in the large set.

# <font color='#FF33A1'>üî• Why Use PCA?</font>
### <font color='red'>üìä Reduce Dimensions:</font> Compress high-dimensional data into fewer dimensions.
### <font color='red'>üéØ Avoid Overfitting:</font> Reduces noise and redundancy.
### <font color='red'>‚ö° Speed Up Models:</font>Less data ‚Üí faster training.

## üìå  Explained Variance Ratio:
        After applying PCA, you can check the amount of variance captured by each principal component.



        pca.explained_variance_ratio_
        
        Sum of explained variances gives the total variance captured by the selected components.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load Dataset
data = load_iris()
X = data.data
y = data.target

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 dimensions
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Explained:", sum(pca.explained_variance_ratio_))

# Plot PCA result
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50)
plt.title('PCA Result (Iris Dataset)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target Labels')
plt.show()
```

## üìä Key Outputs:
        Explained Variance Ratio: Indicates how much variance is captured by each component.
        Scatter Plot: Visualizes high-dimensional data in 2D.


## üî• How Many Principal Components to Select?

        You can use the cumulative variance plot to decide the number of components.


```python
# Plot Cumulative Variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()
plt.show()
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# Here we have reducing the variable into 2d space out of 13
from sklearn.decomposition import PCA
pca=PCA(n_components=2) #Here we are taking 2 independent varibles,those are  PCA defines the whatever the column have highest variance those columns will be selected by the PCA.
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)

# Here It has converted every column into 2d whicher having the highest variance those variance can be shown below

variance=pca.explained_variance_ratio_
```

```python
variance
```

```python
sum(pca.explained_variance_ratio_)
```

        these are our higher variance columns  and thier total sum is 95% that means  new feature having the maximum information from old dataset is 95% and we loosed only 5% data.

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

## üéØ When to Use PCA?
        ‚úÖ High-dimensional data that needs to be compressed.
        ‚úÖ Preprocessing step for machine learning models.
        ‚úÖ Visualizing relationships between high-dimensional points.


## üö® Limitations of PCA:
        1.Linear Assumption: PCA assumes a linear relationship between features.
        2.Loses Interpretability: Principal components are linear combinations of original features, making interpretation difficult.
        3.Sensitive to Scaling: Requires standardization before applying.
