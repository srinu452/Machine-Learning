# <font color='green'>Loss function or Error function or Cost Function<font>

 #####       This difference between the actual values and predicted values is the error 

        A Supervised Machine Learning model aims to train itself on the input variables(X) in such a way that the predicted values(Y) are as closes to the actual values as possible.This difference between the actual values and predicted values is the error and it is used to evaluate the model.The error for any supervised Machine Learning algorithm comprises off 3 Parts:

## <font color='red'>1.BIAS <font>
## <font color='red'>2.Variance Error <font>
## <font color='red'>3.The noise <font>

# <font color='green'>Bias<font>

        Our model will not be trained well with the training data.There will be high training error when we train our model with the data.

        The error introduced by assuming a simple model structure that does not capture the underlying complexity of the data.
        

#### High Bias :
        It indicates that the model is too simple and cannot capture the patterns in the data,leading to "Underfitting".
#### Impact:
            High bias causes the model to have high training error and high test error.

# <font color='green'>Variance<font>

        If you train your data on training data and obtain a very low error,upon changing the data and then training the same previous model you experience high error,this is variance.
         If we train our data ,we got the low train error while testing the  model we get high test error then we say it's variance

         The error introduced by the model being too sensitive to the training data, capturing noise along with the actual patterns.
#### High Variance: 
        Indicates that the model is too complex and overly tuned to the training data, leading to overfitting.
#### Impact: 
            High variance causes the model to perform well on training data but poorly on test data.

### <font color='orange'>Bias-Variance Tradeoff:<font>

        1.The goal in machine learning is to find a balance between bias and variance to minimize the total error (training + test error).
        A model with:
            High bias → Underfitting.
            High variance → Overfitting.

# <font color='green'>Noise<font>

        Noise means ,Our collected data contains any miss values and outliers it leads to decrease our model performance i.e., we can't get good accuracy.

# <font color='orange'>Overfitting<font>

## Defination:
        Overfitting occurs when a model learns the training data too well,including noise and outliers making it performing poorly on new,unseen data.
## Symptoms :
        1.Excellent performance on the training set.
        2.Poor Performance on the test/validation set.
## Causes :
        1.A Model with too many parameters (high complexity)
        2.Insufficient training data.
        3.Noise or irrelevant features in the training data.
## Solutions:
### 1.Regularization: 
        Techniques like L1 (Lasso) or L2 (Ridge) regularization penalize large coefficients to simplify the model.
### 2.Pruning: 
        In decision trees, removing unnecessary branches reduces complexity.
### 3.Early Stopping: 
        Stop training when validation performance stops improving.
### 4.Cross-Validation: 
        Use techniques like k-fold cross-validation to evaluate model performance.
### 5.Simplify the Model: 
        Reduce the number of features or layers in a deep learning model.
### 6.Increase Training Data: 
        Adding more data helps the model generalize better.


        

# <font color='orange'>Under fitting<font>

## 1.Defination : 
        Underfitting occurs when a model is too simple capture the underlying patterns in the data,leading to poor performance on both the training and test sets.
## 2.Symptoms :
        1.High training error 
        2.High test error.
## 3.Causes :
        1.A model with too few parameters (low complexity)
        2.Inadequate training (e.g., insufficient epochs in deep learning).
        3.poor feature selection or representation
## 4.Solutions:
### 1.Increase Model Complexity: 
        Use a more complex model with additional features or layers.
### 2.Train Longer: 
        Allow the model more time to learn the data.
### 3.Feature Engineering: 
        Add or transform features to improve their relevance.
### 4.Decrease Regularization: 
        Reduce penalties on large coefficients if over-regularized.
### 5.Hyperparameter Tuning: 
        Adjust parameters like learning rate or number of layers.
