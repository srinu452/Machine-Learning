# <font color='#FF33A1'>1.Lasso Regression(L1)<font>

        L1 regularization add's the absolute value of the coefficients as a penalty term to the loss function.

        Prevents weights from getting too large (defined by L1 norm).The larger the weights ,the more complex the model is more chances of overfitting.L1 regularization introduces sparsity in the weights .It forces more weights to be zero,than reducing the avarage magnitude of all weights.

# <font color='#FF33A1'>2.Ridge Regression (L2)<font>

        L2 regularization adds the squared value of the coefficients as a penalty term to the loss function.

        Prevents  the weights from getting too large (defined by L2 norm) .The larger the weights ,the more complex the model is ,more chances of overfitting.

```python
Image.open('s84.png')
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1) # alpha =  (Regularization Strength)
ridge.fit(X_train, y_train)
```

# <font color='Darkblue'> Key Takeaways<font>
         L1 Regularization (Lasso)  Shrinks coefficients to zero, good for feature selection.
         L2 Regularization (Ridge)  Shrinks coefficients but keeps them nonzero.
         Elastic Net  A balance of both L1 & L2 for better performance.

# <font color='#FF33A1'> Elastic Net (L1 + L2 Combination)<font>

##### Elastic Net combines both L1 & L2 regularization.

```python
from sklearn.linear_model import ElasticNet

elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)  # l1_ratio = 0.5 means equal L1 and L2
elastic_net.fit(X_train, y_train)
```
