# <font color='green'>Key Concepts in Machine Learning<font>
## 1.Algorithms: 
        These are the mathematical models and methods that process data to make predictions or decisions. Common ML algorithms include linear regression, decision trees, and neural networks.

## 2.Data: 
        ML relies on data for training and testing. The quality, size, and relevance of the data play a significant role in determining the performance of ML models.

## 3.Training and Testing: 
        In ML, data is typically divided into training and testing sets. The model is trained on the training set, learns the patterns, and is then tested on the testing set to evaluate its performance.

## 4.Features: 
        These are the individual measurable properties or characteristics used to make predictions. The process of choosing relevant features is called feature selection, while transforming or creating new features is feature engineering.

## 5.Labels: 
        In supervised learning, labels are the outcomes or the target variable the model tries to predict.

# <font color='orange'>What is Regression ?<font>

        Regression is a Supervised Learning task where output is having continuous value.The goal here is to predict a value as much closer to actual output value as our model can and then evalution is done by calculating error value.The Smaller the error the greater the accuracy of out regression model.

# <font color='orange'>What is Classification?<font>

        Classification is a Supervised Learning task where output is having defined labels(decrete value).For example Output-Purchased has defined labels i.e 0 or 1 means 1 is the customer will purchase and 0 means that the customer won't purchase .It can be either binary or multi-class classification.
        In Binary classification ,the model predicts either 0 or 1 ; yes or no ; but in the case of multi class classification ,model predicts more than one class.

        Example : Gmail classifies mails in more than one classes like social,promotions,updates,offers.
