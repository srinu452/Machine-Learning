# <font color='Darkblue'> Clustering<font>

#### Clustering is an 'Unsupervised Learning Techniques' that groups a set of data points into clusters such that:

#### 1.Data Points in the same cluster are more similar to each other.
#### 2.Data points in different clusters are as dissimilar as possible.


         Clustering Algorithms do the task of dividing the population or data points into a variety of groups such that data points within the same cluster are similar to other data points within the same cluster than those in other groups .Basically ,the aim is to seperate groups with similar traits and assign them into clusters.

# <center> <font color='Darkblue'> K-Means Clustering <font><center>

# ✅ Definition:

        K-Means is a 'Partitioning-based clustering algorithm' that divides data into 'K clusters' by minimizing the 'variance' within each cluster'.

        In this algorithm ,we group the items into K-clusters such that all items in the same cluster are as similar to each other as possible.And items not in the same cluster are as different as possible.

        Distance measures(like Euclidean Distance) are used to calculate similarity and dissimilarity between the data points.Each cluster has a centroid.Centroid can be thought as the point that is most representatively of the cluster.

# <font color='#FF33A1'> ✅ Working of K-Means Algorithm:</font>
#####        1.Choose the number of clusters (K).
        You decide how many groups you want to create.
        
#####        2.Initialize K centroids randomly.
        
#####        3.Assign each data point to the nearest centroid (forming K clusters).
        1.Measure distance (usually Euclidean distance).
        2.Each point is assigned to the closest cluster center.
        
#####        4.Recompute centroids as the mean of points in each cluster.
        After all points are assigned, recalculate the centroid (mean of all points in that cluster).
        
#####        5.Repeat steps 3 and 4 until centroids do not change (convergence).
        1.Centroids no longer change (convergence), or
        2.A maximum number of iterations is reached.

# <center><font color='Darkblue'> Implementation of K-Means Clustering</font></center>

# ✅ 1. K-Means Clustering Implementation (from Scratch)

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
X = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [10, 2],
    [10, 4],
    [10, 0]
])

# Number of clusters
K = 2

# Randomly initialize centroids
centroids = X[np.random.choice(X.shape[0], K, replace=False)]
print("Initial Centroids:\n", centroids)

# Function to compute distance
def distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

# K-Means Algorithm
for _ in range(10):  # Number of iterations
    # Step 1: Assign clusters
    clusters = [[] for _ in range(K)]
    for point in X:
        distances = [distance(point, centroid) for centroid in centroids]
        cluster_idx = np.argmin(distances)
        clusters[cluster_idx].append(point)
    
    # Step 2: Update centroids
    new_centroids = []
    for cluster in clusters:
        new_centroids.append(np.mean(cluster, axis=0))
    centroids = np.array(new_centroids)
    print("Updated Centroids:\n", centroids)

# Final Output
print("Final Centroids:\n", centroids)

# Plotting
for cluster in clusters:
    cluster = np.array(cluster)
    plt.scatter(cluster[:, 0], cluster[:, 1])
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', s=100)
plt.title("K-Means Clustering Result")
plt.show()
```

## <font color='Darkblue'>Importing the Required Libraries<font>

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# <font color='Darkblue'> Data Preprocessing</font>

```python
dataset.isnull().sum()
```

```python
X=dataset.iloc[:,3:].values
X
```

```python
# TO know the k value we are calculating elbow method 
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):# here we are taking 10 clusters
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # It will calculate the distance between cluster point and data points near tp them
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel('No of clusters')
plt.ylabel("WCss")
plt.show()
```

        Here we got the k=5 where we got shape as Elbow there we have to considered that is our optimization K value

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
plt.scatter(X[y_kmeans==0, 0],X[y_kmeans==0,1], s=100 ,c='red',label='Cluster 1')
plt.scatter(X[y_kmeans==1, 0],X[y_kmeans==1,1], s=100 ,c='blue',label='Cluster 2')
plt.scatter(X[y_kmeans==2, 0],X[y_kmeans==2,1], s=100 ,c='orange',label='Cluster 3')
plt.scatter(X[y_kmeans==3, 0],X[y_kmeans==3,1], s=100 ,c='violet',label='Cluster 4')
plt.scatter(X[y_kmeans==4, 0],X[y_kmeans==4,1], s=100 ,c='pink',label='Cluster 5')
plt.xlabel('Annual_income')
plt.ylabel('Spending_score')
plt.show()
```
