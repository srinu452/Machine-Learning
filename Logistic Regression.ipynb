## <font color='green'>Key Concepts in Logistic Regression :<font>

#### 1.Binary Outcome: 
        Logistic regression is most commonly used when the target variable is binary (two classes). It can also be extended to multiclass classification.

#### 2.Sigmoid Function: 
            Logistic regression uses the sigmoid function (also known as the logistic function) to transform the linear output into a probability between 0 and 1. This probability can then be mapped to class labels (e.g., class 1 if P0.5, else class 0).

#### 3.Log Odds: 
           Logistic regression models the probability of a particular outcome as a function of the input variables using a log-odds transformation. This allows it to handle binary classification problems by computing probabilities.

## <font color='green'>Types of Logistic Regression<font>

#### 1.Binary Logistic Regression : 
        The target variable has only two possible outcomes.For example,classifying e-mails as Spam or not Spam.

#### 2.Multinomial Logistic Regression :
        The taget Variable has three or more categories without ordering.For example,predicting which food is prefferd more(Veg,Non-Veg,Vegan)
#### 3.Ordinal Logistic Regression : 
        The target variable has three or more categories with ordering .For example,rating for a movie rating from 1 to 5.

## <font color='green'>Cost Function in Logistic Regression<font>

        Logistic regression uses a log loss(binary cross-entropy) cost function,which measures the accuracy of the classifier.The goal of training is to minimize this cost function.

## <font color='green'>Logistic Regression Example: Predicting Disease Presence<font>
Imagine using logistic regression to predict whether a patient has a certain disease (yes or no).

#### 1.Features: 
        Age, blood pressure, cholesterol levels, etc.
#### 2.Target Variable: 
        Disease presence (1) or absence (0).
#### 3.Model: 
        Logistic regression builds a model using these features to predict the probability of disease presence.
#### 4.Interpretation:
        The output probability (e.g., 0.8) indicates an 80% chance that the patient has the disease. Based on a threshold (often 0.5), you can classify this as "disease" if the probability is above 0.5.

## <font color='green'>Advantages:<font>

#### 1.Simple and Interpretable: 
        Outputs probabilities, making it easy to interpret.
#### 2.Fast and Efficient: 
        Works well on small to medium-sized datasets.
#### 3.Well-Suited for Binary Classification: 
        Excellent for binary outcomes and straightforward to implement.

## <font color='green'>Disadvantages:<font>

#### 1.Not Effective for Complex Relationships: 
        Assumes a linear relationship between features and the log-odds, so it may struggle with non-linear relationships.
#### 2.Sensitive to Imbalanced Data:
        Performance can be affected by imbalance in the classes (e.g., many more negatives than positives).

## <font color='green'>Applications:<font>
        Spam detection, medical diagnosis, loan approval, and binary decision-making.

```python
# creating feature matric and dependent variable vector
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
#splitting the data into training data and test data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
## for getting better accuracy we uses feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train[0:10]
```

```python
X_test[0:10]
```

## <font color='blue'>Training the Model<font> 

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,Y_train)
```

## <font color='blue'>Testing or Predicting the Model<font>

```python
Y_pred=classifier.predict(X_test)
Y_pred
```

## <font color='blue'>Confusion Matrix and Accuracy<font>

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,Y_pred)
cm
```

```python
sns.heatmap(cm, 
            annot=True,
            fmt='g', 
            xticklabels=[0,1],
            yticklabels=[0,1])
plt.ylabel('Actual', fontsize=13)
plt.title('Confusion Matrix', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.xlabel('Prediction', fontsize=13)
plt.gca().xaxis.tick_top()

plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Prediction', ha='center', fontsize=13)
plt.show()
plt.show()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(Y_test,Y_pred))
confusion_matrix(Y_test,Y_pred)
```

## <font color='blue'>Classifications Report based on Confusion Metrics<font>

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```
